# 03 — Verdade-terreno das relações espaciais

Este notebook explora as tabelas de verdade-terreno geradas por `src/relations_ground_truth.py`:

- `ground_truth_unarios.csv` — predicados unários (isRed, isCircle, isSmall...);
- `ground_truth_pares.csv` — predicados binários, 25×25 = 625 pares;
- `ground_truth_triplas.csv` — predicado ternário `inBetween`, 25×24×23 = 13.800 triplas.

```python
from pathlib import Path
import json

import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import Image, Markdown, display


def find_project_root(start: Path) -> Path:
    """Localiza a raiz do projeto subindo diretórios até achar src/ e requirements.txt."""
    current = start.resolve()
    for candidate in [current, *current.parents]:
        if (candidate / "requirements.txt").exists() and (candidate / "src").exists():
            return candidate
    raise FileNotFoundError("Não foi possível localizar a raiz do projeto.")


PROJECT_ROOT = find_project_root(Path.cwd())
DATA_DIR = PROJECT_ROOT / "data" / "datasets_gerados"
RESULTS_DIR = PROJECT_ROOT / "results"
METRICS_DIR = RESULTS_DIR / "metrics"
QUERIES_DIR = RESULTS_DIR / "queries"
TRAINING_DIR = RESULTS_DIR / "training"

print("Raiz do projeto:", PROJECT_ROOT)
```

## Predicados unários

```python
df_unarios = pd.read_csv(DATA_DIR / "ground_truth_unarios.csv")

df_unarios.groupby("predicado")["y_true"].sum().sort_values(ascending=False)
```

## Predicados binários

As relações espaciais binárias usam limiares fixos definidos em `relations_ground_truth.py`:

- `DEFAULT_CLOSE_THRESHOLD = 0.25` (para `closeTo`);
- `DEFAULT_STABLE_HORIZONTAL_THRESHOLD = 0.12` (para `canStack`);
- `DEFAULT_SEGMENT_TOLERANCE = 0.05` (para `inBetween` via segmento de reta).

```python
df_pares = pd.read_csv(DATA_DIR / "ground_truth_pares.csv")

colunas_relacoes = ["leftOf", "rightOf", "below", "above", "closeTo",
                    "sameColor", "sameShape", "sameSize", "canStack"]

df_pares[colunas_relacoes].mean().rename("proporção de positivos")
```

```python
# Exemplo: pares onde closeTo é verdadeiro
df_pares[df_pares["closeTo"] == 1].head(10)
```
